```python
def quicksort(A, left, right, steps):
    #print(f"Aufruf mit {A[left]}, {A[right-1]}")
    if right - left <= 1:
        return  # Basisfall: Ein- oder kein Element
    
    pivot_index = left # Pivot ist das erste Element
    pivot = A[pivot_index]
    
    lmbda = left + 1  # Start direkt nach dem Pivot
    rho = right - 1
    #print(f"pi: {pivot_index}, pivot: {A[pivot_index]}")
    while lmbda <= rho:
        
        #print(f"lambda: {lmbda}, rho: {rho}")
        while lmbda <= rho and A[lmbda] < pivot:
            lmbda += 1
        while lmbda <= rho and A[rho] > pivot:
            rho -= 1
        
        if lmbda < rho:
            #print(f"Vertauschung bei: lambda {lmbda}, rho {rho} bzw. {A[lmbda]},{A[rho]}")
            A[lmbda], A[rho] = A[rho], A[lmbda]
            lmbda += 1
            rho -= 1
            #print(f"Vertauschung: {A}")
            #plt.plot(range(0, len(A)), A)
            #plt.show()
            steps.append(A.copy())
    
        
    
    # Platzieren des Pivots an die richtige Position
    A[left], A[rho] = A[rho], A[left]  # rho ist jetzt der Pivot-Platz
    if left != rho: 
        #print(f"Pivotelement platziert: {A}")
        #plt.plot(range(0, len(A)), A)
        #plt.show()
        steps.append(A.copy())
    
    # Rekursion
    quicksort(A, left, rho, steps)  # Linke Hälfte
    quicksort(A, rho + 1, right, steps)  # Rechte Hälfte

# Beispielaufruf
arr = [4, 5, 3, 9, 7, 1, 2, 8, 6]
steps = []
quicksort(arr, 0, len(arr), steps)
print(arr)  # Erwartete Ausgabe: [1, 2, 3, 4, 5, 6, 7, 8, 9]
```

```python
#starts with the rightmost element
def quicksort(A, left, right, steps):
    if right - left <= 1:
        return  # Basisfall: Ein- oder kein Element
    
    pivot_index = right - 1  # Letztes Element als Pivot
    pivot = A[pivot_index]
    
    lmbda = left  # Links startet beim ersten Element
    rho = right - 2  # Rechts startet vor dem Pivot

    while lmbda <= rho:
        while lmbda <= rho and A[lmbda] < pivot:
            lmbda += 1
        while lmbda <= rho and A[rho] > pivot:
            rho -= 1

        if lmbda < rho:
            A[lmbda], A[rho] = A[rho], A[lmbda]
            lmbda += 1
            rho -= 1
            steps.append(A.copy())  # Schritt speichern

    # Platzieren des Pivots an die richtige Position
    A[right - 1], A[lmbda] = A[lmbda], A[right - 1]
    steps.append(A.copy())  # Schritt speichern

    # Rekursive Aufrufe für linke und rechte Partition
    quicksort(A, left, lmbda, steps)   # Linke Hälfte
    quicksort(A, lmbda + 1, right, steps)  # Rechte Hälfte

# Beispielaufruf
arr = [4, 5, 3, 9, 7, 1, 2, 8, 6]
steps = []
quicksort(arr, 0, len(arr), steps)
print(arr)  # Erwartete Ausgabe: [1, 2, 3, 4, 5, 6, 7, 8, 9]
```

```python
def quicksort(A, left, right, steps, pivot_selector):
    if right - left <= 1:
        return  # Base case: One or zero elements
    
    pivot_index = pivot_selector(A, left, right)  # Get pivot index from function
    A[pivot_index], A[right - 1] = A[right - 1], A[pivot_index]  # Move pivot to end
    pivot = A[right - 1]

    lmbda = left  # Left pointer
    rho = right - 2  # Right pointer (before pivot)

    while lmbda <= rho:
        while lmbda <= rho and A[lmbda] < pivot:
            lmbda += 1
        while lmbda <= rho and A[rho] > pivot:
            rho -= 1

        if lmbda < rho:
            A[lmbda], A[rho] = A[rho], A[lmbda]
            lmbda += 1
            rho -= 1
            steps.append(A.copy())  # Save step

    # Place pivot at correct position
    A[right - 1], A[lmbda] = A[lmbda], A[right - 1]
    steps.append(A.copy())

    # Recursive calls for left and right partitions
    quicksort(A, left, lmbda, steps, pivot_selector)   # Left part
    quicksort(A, lmbda + 1, right, steps, pivot_selector)  # Right part

# Example pivot selection strategies:
def choose_first(A, left, right):
    return left

def choose_last(A, left, right):
    return right - 1

def choose_middle(A, left, right):
    return (left + right - 1) // 2

def choose_random(A, left, right):
    import random
    return random.randint(left, right - 1)

# Example call:
arr = [4, 5, 3, 9, 7, 1, 2, 8, 6]
steps = []
quicksort(arr, 0, len(arr), steps, choose_middle)  # Change pivot selection function here
print(arr)  # Expected output: [1, 2, 3, 4, 5, 6, 7, 8, 9]
```
